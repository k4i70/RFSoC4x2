```python
'''
This is some code I wrote that works with my custon overlay to generate a different waveform on each DAC
'''

import numpy as np
import matplotlib.pyplot as plt
import time
import xrfdc
from scipy.signal import sawtooth, square, butter, lfilter
from mulitple_output import SeparateDacOverlay
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import chirp, sawtooth


# This reads the overlay that is created by the Vivado project
ol = SeparateDacOverlay('./Working_Overlays/multi_dac_4gsps.bit')
```

```python
DAC_SR = 4.0e9      # 4 GS/s
A = 2**15-1           # amplitude limit for int16 since sine and square are 2 PP, and -1 to account for 0.
N = ol.dac0_bram.shape[0]   # BRAM length
t = np.arange(N) / DAC_SR
```

```python
f0 = 10e6    # 50 MHz for the sine wave
f1 = 10e6 # 10 MHz for the square wave

# 1. Sine wave on channel 0
# wave0 = A * np.sin(2*np.pi*f0*t)
wave0 = A * square(2 * np.pi * f1 * t)

# 2. Square wave on channel 1
# wave1 = A * sawtooth(2 * np.pi * f0_sin * t, width=0.0)
wave1 = A * np.sin(2 * np.pi * f1 * t)*0
# wave1 = A * np.sin(2*np.pi*f0*t)
```

```python
wave0_i16 = wave0.astype(np.int16)
wave1_i16 = wave1.astype(np.int16)

ol.dac0_bram[:] = wave0_i16
ol.dac1_bram[:] = wave1_i16
print("Loaded waveforms into both DAC BRAMs.")
# This is what starts the DACs outputting the waveforms
ol.start_dacs()

# ol.trigger_capture()
```

```python
# There is probably  better way to turn it off, but I just set the outputs to 0.
ol.dac0_bram[:] = wave0_i16*0
ol.dac1_bram[:] = wave1_i16*0
ol.start_dacs()
```
